```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2020_Predictions.csv', index_col=0)
```

```python
df
```

```python
df.columns
```

```python
obvious_drop = ['player_id', 'attempts', 'Rk', 'year', 'pa', 'bip', 'est_ba_minus_ba_diff', 
                
                'est_slg_minus_slg_diff', 'est_woba_minus_woba_diff', 'Lg', 'G', 'AB', 'R', 'H', '2B', '3B', 
                'BB', 'SB', 'CS', 'OPS+', 'TB', 'GDP', 'HBP', 'Pos Summary', 'ev95plus', 'barrels', 'brl_percent']
```

```python
df1 = df.drop(obvious_drop, axis=1)
```

```python
df1['RBI/PA'] = df1['RBI'] / df1['PA▼']

df1['SO/PA'] = df1['SO'] / df1['PA▼']

df1['SH/PA'] = df1['SH'] / df1['PA▼']

df1['SF/PA'] = df1['SF'] / df1['PA▼']

df1['IBB/PA'] = df1['IBB'] / df1['PA▼']

df1['CHR_per_PA'] = df1['HR'] / df1['PA▼']

df1.drop(['RBI', 'SO', 'SH', 'SF', 'IBB'], axis = 1, inplace = True)
```

```python
df1.columns
```

```python
in_depth_drop_list = ['max_hit_speed', 'gb', 'max_distance', 'avg_distance', 'BA', 'SLG']
```

```python
df2 = df1.drop(in_depth_drop_list, axis = 1)
```

```python
#create angle sweet spot index. 1 for less than 28 degrees, 2 for between 28 and 35 degrees, 3 for more than 35 degrees
df2['angle_sweet_spot_index'] = df2['anglesweetspotpercent'].apply(lambda x: 1 if x < 28 else ( 2 if 28 <= x <= 35 else 3))
```

```python
#create dummy variables for angle sweet spot index
sweet_spot_dummies = pd.get_dummies(df2.angle_sweet_spot_index, drop_first=True, prefix='AS_')
```

```python
df3 = pd.concat([df2, sweet_spot_dummies], axis=1)
```

```python
df3.columns
```

```python
df3.to_csv('2020_prediction_dataset.csv')
```
